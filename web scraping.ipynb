```python
from bs4 import BeautifulSoup
html = """
<html>
    <body>
        <h1 id="title">웹 스크레이핑이란?</h1>
        <p id="body">웹 페이지 분석을 통해</p>
        <p id="body">원하는 내용을 끄집어 냄</p>
    </body>
</html>
"""

soup = BeautifulSoup(html, 'html.parser')

bodys = soup.find_all(id="body")

for p in bodys:
    id1 = p.sttrs
    text = p.string
    id2 = "body"
    print(id1, id2,  " = ", text )
```
