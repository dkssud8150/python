```python
from bs4 import BeautifulSoup

url2 = "http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108"

res = urllib.request.urlopen(url2)
```

```python
soup = BeautifulSoup(res, "html.parser")
```

```python
def check_null(tag, x):
    if x is None:
        print(tag, '값이 없습니다.')
    else:
        print(x, x.getText())
```

```python
type(soup.find("title"))
```

```python
check_null('title', soup.find('title')) 
#soup.find('link').getText()
#check_null('link', soup.find('link'))
check_null('pubDate', soup.find('pubDate'))
check_null("wf", soup.find("wf"))
```
